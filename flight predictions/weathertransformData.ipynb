```python
import pandas as pd
```

```python
import os
def process_weather_data(file_path):
    flight_data = pd.read_csv(file_path)
    # Convert the timestamp to a pandas datetime object
    datetime_obj = pd.to_datetime(flight_data['timestamp_local'])

    # Filter data for years 2022 and 2023
    mask = (datetime_obj.dt.year == 2022) | (datetime_obj.dt.year == 2023)
    flight_data = flight_data[mask]
    
    # Extract the airport code from the file name
    airport_code = os.path.basename(file_path).split('_')[0].upper()

    # Extract the hour and minute components
    flight_data['Date'] = datetime_obj.dt.strftime('%m/%d/%Y')
    #pred_data['Date'] = datetime_obj.dt.strftime('%m/%d/%Y')
    flight_data['hours'] = datetime_obj.dt.hour
    flight_data['minutes'] = datetime_obj.dt.minute

    columns_to_drop = ['app_temp', 'dhi', 'dni', 'ghi', 'pod', 'slp', 'solar_rad', 'timestamp_local', 'timestamp_utc', 'ts', 'uv', 'weather.icon']
    flight_data.drop(columns=columns_to_drop, inplace=True)
    file_name = os.path.basename(file_path)
    if 'syr' in file_name.lower():
        # flight_data = flight_data.rename(columns={'minutes': 'minutes', 'hour': 'hours'})
        flight_data.columns = flight_data.columns.map(lambda x: 'arr_' + str(x))
        flight_data.rename(columns={'arr_Date': 'Date'}, inplace=True)
    else:
        # flight_data = flight_data.rename(columns={'minutes': 'dep_minutes', 'hour': 'dep_hours'})
        flight_data['Origin_Airport'] = airport_code
        flight_data.columns = flight_data.columns.map(lambda x: 'dep_' + str(x))
        flight_data.rename(columns={'dep_Date': 'Date', 'dep_Origin_Airport': 'Origin_Airport'}, inplace=True)
    
    return flight_data
```

```python
# List of file paths

#file_paths = ['data_miner/data/SYR_weather_data_forecast.csv', 'data_miner/data/JFK_weather_data_forecast.csv', 
#              'data_miner/data/MCO_weather_data_forecast.csv', 'data_miner/data/ORD_weather_data_forecast.csv']
file_paths = ['data_miner/data/SYR_weather_data_hourly.csv', 'data_miner/data/JFK_weather_data_hourly.csv', 
              'data_miner/data/MCO_weather_data_hourly.csv', 'data_miner/data/ORD_weather_data_hourly.csv']

for file in file_paths:
    processed_data = process_weather_data(file)
    file_name = os.path.basename(file)
    file_name_no_ext, file_ext = os.path.splitext(file_name)
    new_file_name = f'{file_name_no_ext}_processed.csv'
    print(processed_data.head())
    processed_data.to_csv(f'weather_data/{new_file_name}', index=False)
    print(len(processed_data.columns))
```
